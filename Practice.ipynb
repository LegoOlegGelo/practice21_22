## Partial limits

### №117

```python
def full_seq(n):
    return [1/i for i in range(1, (n+1) + 1)]


print('Последовательность:')
for i in range(10):
    print(*full_seq(i), sep=', ')
```

```python
def part_of_seq(n):
    return 1 + 1/n, 1/2 + 1/n, 1/(n-1) + 1/n, 1/(n+1)


for i in range(1, 10):
    p = part_of_seq(10**i)
    print(*p[:2], '...', *p[2:], sep=', ')
```

Как видно, при увеличении n, члены последовательности стремятся к $ 1, \frac{1}{2}, \frac{1}{3}, ..., 0 $

### №120

```python
def get_nth(n):
    a, b = sp.symbols('a b')
    return 1/2 * ((a + b) + (-1)**n * (a - b))


for i in range(10):
    print(get_nth(i), end=', ')

print('\n')
    
for i in range(1, 6):
    print((get_nth(10**i), get_nth(10**i + 1)))
```

## 2. Решить уравнение методом итераций

Написать программу на python для поиска приближенного решения уравнения $ \sqrt{1 - 0.4x^2} - arcsinx = 0 $ на отрезке [0;1] методом итераций. Точность вычисления решения $ 10^{-5} $.

Преобразуем уравнение к виду x = g(x):

$ \sqrt{1 - 0.4x^2} - arcsinx = 0 $

$ arcsinx = \sqrt{1 - 0.4x^2} $

$ sin(arcsinx) = sin(\sqrt{1 - 0.4x^2}) $

$ x = sin(\sqrt{1 - 0.4x^2}) $

```python
x = sp.Symbol('x')
expr =  sp.sin(sp.sqrt(1 - 0.4*x**2))
df = sp.lambdify(x, sp.diff(expr))
X = np.linspace(0, 1, 200)
Y = df(X)

plt.plot(X, Y, label='df/dx')
_ = plt.legend()
```

Модуль производной меньше 1 на отрезке [0;1]

```python
def f(x):
    return math.sin(math.sqrt(1 - 0.4*x**2))


def find(f, x, eps=10e-5):
    it = 0
    while True:
        old_x = x
        x = f(x)
        it += 1
        if abs(old_x - x) < eps or it >= 20_000:
            return x
        

find(f, 1/2)
```

0.7671683986506608 - приближенное значение (с точностью $ 10^{-5} $) корня уравнения на отрезке [0;1]

```python
# тестирование функции при различных начальных значениях

for v in np.linspace(0, 1, 30):
    print(f'[x0 = {v}]: {find(f, v)}')
```

Как видно, корень находится вне зависимости от выбора начального значения на отрезке [0;1]. Все значения имеют вид 0.7671\*\*\*. Точность в $ 10^{-5} $ достигнута.
