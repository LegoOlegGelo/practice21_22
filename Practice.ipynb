```python
import numpy as np
import math

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def limit(f, x0, sign=1):
    """f - function, x0 - limit point, sign=1 - right limit, sign=-1 - left limit"""
    dx_list = [1, 0.1, 10e-2, 10e-3, 10e-4, 10e-5, 10e-6]
    return [f(x0 + sign*dx) for dx in dx_list]
```

## Functions plots

### 259
   $ y = \frac{x}{1-x^2} $

```python
def f(x):
    return x / (1 - x**2)

X = np.linspace(-5, 5, 200)
Y = f(X)

plt.plot(X, Y, label='y = x / (1 - x^2)')
_ = plt.legend()
```

### 285
   $ y = sin(x - x_0),  \hspace{1mm} x_0 = 0, \frac{\pi}{4}, \frac{\pi}{2}, \frac{3\pi}{4}, \pi  $

```python
def f(x, x0):
    return np.sin(x - x0)


X = np.linspace(-5, 5, 200)

for i in range(4):
    Y = f(X, i*np.pi/4)
    plt.plot(X, Y, label=f'y = sin(x - {i}*pi/4)')


_ = plt.legend()
```

### 292
   $ y = \frac{1}{sinx} $

```python
def f(x):
    return 1 / np.sin(x)

X = np.linspace(-5, 5, 1000)
Y = f(X)

plt.plot(X, Y, label='y = csc(x)')
plt.ylim((-10, 10))
_ = plt.legend()
```

## Limits

### 553

 $ \lim_{x \to +\infty}{(\sqrt[n]{x} - \sqrt[n+1]{x})}  \hspace{4mm}, x > 0 $

```python
def f(x, n=2):
    return n**2 * (x**(1/n) - x**(1/(n+1)))


for i in range(0, 100, 8):
    print(f(10**i))
```

$ .. (\sqrt[n]{x} - \sqrt[n+1]{x}) \to +\infty,  \hspace{1mm} x \to +\infty,  \\ $ 
$ \lim_{x \to +\infty}{(\sqrt[n]{x} - \sqrt[n+1]{x})} = +\infty $

### 554

 $ \lim_{n \to +\infty}{\left(\frac{a - 1 + \sqrt[n]{b}}{a} \right)^n}  \hspace{4mm}, a > 0, b > 0 $

```python
def f(n, a=2, b=25):
    return ((a - 1 + b**(1/n)) / a)**n


for i in range(0, 20):
    print(f(10**i))
```

 15   ,     $ \sqrt[a]{b} $

  -   float  $ \frac{1}{n} $  $ n > 10^{14} $   , -      - 1.

$ \lim_{n \to +\infty}{(\frac{a - 1 + \sqrt[n]{b}}{a})^n} = \sqrt[a]{b} $

### 555

 $ \lim_{n \to +\infty}{\left( \frac{\sqrt[n]{a} + \sqrt[n]{b}}{2} \right) ^n}  \hspace{4mm}, a > 0, b > 0 $

```python
def f(n, a=4, b=25):
    return ((a**(1/n) + b**(1/n)) / 2)**n


for i in range(0, 20):
    print(f(10**i))
```

### 556

 $ \lim_{x \to 0}{\left(\frac{a^x + b^x + c^x}{3} \right)^\frac{1}{x}}  \hspace{4mm}, a > 0, b > 0, c > 0 $

```python
def f(x, a=8, b=27, c=64):
    return ((a**x + b**x + c**x) / 3)**(1/x)


for i in range(15):
    print(f(0 + 10**-i))
```

   24 = $ \sqrt[3] a \sqrt[3] b \sqrt[3] c $, ..

$ \lim_{x \to 0}{\left(\frac{a^x + b^x + c^x}{3} \right)^\frac{1}{x}} = \sqrt[3] a \sqrt[3] b \sqrt[3] c $

### 557

```python
def f(x, a=2, b=3, c=4):
    return ((a**(x + 1) + b**(x + 1) + c**(x + 1)) / (a + b + c))**(1/x)


for i in range(15):
    print(f(0 + 10**-i))
```
