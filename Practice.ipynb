```python
def full_seq(n):
    return [1/i for i in range(1, (n+1) + 1)]


print(':')
for i in range(10):
    print(*full_seq(i), sep=', ')
```

## Derivatives

```python
x = sp.symbols('x')
expr = sp.exp(x) * sp.sin(x)


def nth_deriv(expr, n=1):
    """ Function of nth derivative """
    if n < 1: return expr
    return nth_deriv(sp.diff(expr), n - 1)


for i in range(10):
    print(f'({i}):', nth_deriv(expr, i))
```

```python
def f(x):
    return math.sin(math.sqrt(1 - 0.4*x**2))


def find(f, x, eps=10e-5):
    it = 0
    while True:
        old_x = x
        x = f(x)
        it += 1
        if abs(old_x - x) < eps or it >= 20_000:
            return x
        

find(f, 1/2)
```
