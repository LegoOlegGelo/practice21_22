## Partial limits

### №117

```python
def full_seq(n):
    return [1/i for i in range(1, (n+1) + 1)]


print('Последовательность:')
for i in range(10):
    print(*full_seq(i), sep=', ')
```

```python
def part_of_seq(n):
    return 1 + 1/n, 1/2 + 1/n, 1/(n-1) + 1/n, 1/(n+1)


for i in range(1, 10):
    p = part_of_seq(10**i)
    print(*p[:2], '...', *p[2:], sep=', ')
```

Как видно, при увеличении n, члены последовательности стремятся к $ 1, \frac{1}{2}, \frac{1}{3}, ..., 0 $

### №118

```python
def full_seq(n):
    return [i/n for i in range(1, n)]


print('Последовательность:')
for i in range(10):
    print(*full_seq(i), sep=', ')
```

```python
def part_of_seq(n):
    return 1/n, 2/n, 3/n, (n-3)/n, (n-2)/n, (n-1)/n


for i in range(1, 10):
    p = part_of_seq(10**i)
    print(*p[:3], '...', *p[3:], sep=', ')
```

Как видно, при увеличении n, члены последовательности начинают принимать все значения из [0;1]

### №119

```python
def get_nth(n):
    return 3 * (1 - 1/n) + 2 * (-1)**n


print('Последовательность:')
print(*[round(get_nth(i), 2) for i in range(1, 200)])
```

```python
print(*[(get_nth(10**i), get_nth(10**i + 1)) for i in range(5, 15)], sep='\n')
```

Как видно, при увеличении n, члены последовательности стремятся к 5 (четные) и 1 (нечетные).

### №120

```python
def get_nth(n):
    a, b = sp.symbols('a b')
    return 1/2 * ((a + b) + (-1)**n * (a - b))


for i in range(10):
    print(get_nth(i), end=', ')

print('\n')
    
for i in range(1, 6):
    print((get_nth(10**i), get_nth(10**i + 1)))
```

Как видно, при увеличении n, члены последовательности стремятся к a (четные) и b (нечетные).

## Functions plots

### №259
Построить график функции $ y = \frac{x}{1-x^2} $

```python
def f(x):
    return x / (1 - x**2)

X = np.linspace(-5, 5, 200)
eps = 10e-2
X[np.abs(X**2 - 1) < eps] = np.nan
Y = f(X)

plt.plot(X, Y, label='y = x / (1 - x^2)')
_ = plt.legend()
```

## Derivatives

```python
x = sp.symbols('x')
expr = sp.exp(x) * sp.sin(x)


def nth_deriv(expr, n=1):
    """ Function of nth derivative """
    if n < 1: return expr
    return nth_deriv(sp.diff(expr), n - 1)


for i in range(10):
    print(f'({i}):', nth_deriv(expr, i))
```

```python
# ex 3003

n, x = sp.symbols('n x')
expr = ((-1)**n * n**3) * x**n / (sp.factorial(n + 1))
res = sp.Sum(expr, (n, 0, float('+inf'))).doit()
sp.simplify(res)
```
