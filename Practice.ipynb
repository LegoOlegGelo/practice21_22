```python
import numpy as np
import math
import sympy as sp

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def limit(f, x0, sign=1):
    """f - function, x0 - limit point, sign=1 - right limit, sign=-1 - left limit"""
    dx_list = [1, 0.1, 10e-2, 10e-3, 10e-4, 10e-5, 10e-6]
    return [f(x0 + sign*dx) for dx in dx_list]
```

### 259
   $ y = \frac{x}{1-x^2} $

```python
def f(x):
    return x / (1 - x**2)

X = np.linspace(-5, 5, 200)
eps = 10e-2
X[np.abs(X**2 - 1) < eps] = np.nan
Y = f(X)

plt.plot(X, Y, label='y = x / (1 - x^2)')
_ = plt.legend()
```

### 285
   $ y = sin(x - x_0),  \hspace{1mm} x_0 = 0, \frac{\pi}{4}, \frac{\pi}{2}, \frac{3\pi}{4}, \pi  $

```python
def f(x, x0):
    return np.sin(x - x0)


X = np.linspace(-5, 5, 200)

for i in range(4):
    Y = f(X, i*np.pi/4)
    plt.plot(X, Y, label=f'y = sin(x - {i}*pi/4)')


_ = plt.legend()
```

## Derivatives

```python
x = sp.symbols('x')
expr = sp.exp(x) * sp.sin(x)


def nth_deriv(expr, n=1):
    """ Function of nth derivative """
    if n < 1: return expr
    return nth_deriv(sp.diff(expr), n - 1)


for i in range(10):
    print(f'({i}):', nth_deriv(expr, i))
```
